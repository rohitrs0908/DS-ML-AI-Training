Sales Prediction: The Dataset used in these models tells about whether a person of certain age having certain income purchases a product or not.
We need to predict whether a targeted audience will purchase the product or not.

```python
# import the library
import pandas as pd
# load the dataset
dataset = pd.read_csv('Social_Network_Ads.csv')
```

```python
# display the first five rows
dataset.head()
```

```python
# check the null values
dataset.isnull().sum()
```

```python
# ID is not helpful in predicting , so we will drop ID
dataset.drop(columns=['User ID'],inplace = True)
```

```python
# Define X and y
# X is independent variable/features (Gender , Age and Estimated Salary)
# y is depencdent variable/label/target (Purchased)
X = dataset.iloc[:,0:3]
y = dataset.iloc[:,3]
```

```python
# Character values in gender to be converted to numeric
# import the library
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() # instantiating object of Labelencoder class.
X['Gender'] = le.fit_transform(X['Gender'])
X.head()
```

```python
# Split into train and test
# train will be used for training and test for testing
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X, y , test_size=0.2 , random_state=5)
```

```python
dataset.describe()
```

```python
dataset.info()
```

```python
# We need to standardise the values
# import the library
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Build the model
#import the library
import tensorflow as tf
from keras.layers import *
from keras.models import *
```

```python
model = Sequential()
# input layer
model.add(Dense(16,input_dim=3,activation='relu'))
# hidden layer
model.add(Dense(16,activation='relu'))
model.add(Flatten())
# output layer
model.add(Dense(1 , activation= 'sigmoid'))
model.summary()
```

```python
# Compile the model
model.compile(loss='binary_crossentropy',optimizer='adam',metrics='accuracy')
```

```python
# use fit to train the model
model.fit(X_train,y_train,batch_size=16,epochs=10)
```

```python
# predict the values
y_pred = model.predict(X_test)
y_pred = (y_pred>0.5)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```
