```python
# import pandas and sklearn's CountVectorizer class
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# set of documents
corpora = ['The quick brown fox.','Jumps over the lazy dog!']
```

```python
# instantiate the vectorizer object
#cvec = CountVectorizer(lowercase=False)
cvec = CountVectorizer()
# convert the documents into a document-term matrix
wm = cvec.fit_transform(corpora)

# retrieve the terms found in the corpora
tokens = cvec.get_feature_names()
```

```python
pd.DataFrame(data=wm.toarray(), index=['Doc1', 'Doc2'],
             columns=tokens)
```

```python
cvec.vocabulary_
```

```python
pd.DataFrame({'Words': cvec.vocabulary_.keys(),'Index' : cvec.vocabulary_.values()},index=cvec.vocabulary_.values())
```
